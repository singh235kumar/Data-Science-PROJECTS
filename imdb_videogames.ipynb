```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
raw_df=pd.read_csv('imdb(VIDEO GAMES).csv')
```

```python
raw_df
```

```python
to_drop = ['certificate', 'DETAILS', 'ACTOR-1', 'ACTOR-2', 'ACTOR-3', 'ACTOR-4']
movies_df = raw_df.drop(columns= to_drop)
```

```python
movies_df.head()
```

```python
movies_df.isnull()
```

```python
movies_df=movies_df.fillna(0)
```

```python
to_rename = {
    
    'VIDEO GAME NAME ': 'name',
    'Year': 'year',
    'DIRECTOR ': 'director',
    
}
movies_df.rename(columns=to_rename,inplace=True)
```

```python
movies_df
```

```python

toop_per_rat=movies_df[['name','Rating']].sort_values('Rating',ascending=False).head(10)
```

```python

plt.title('top 10 best video games')
sns.barplot(x='Rating',y='name',data=toop_per_rat);
```

```python
low_per_rat=movies_df[['name','Rating']].sort_values('Rating',ascending=True).head(10)
```

```python
low_per_rat
```

```python
plt.title('TOP 10 WORST VIDEO GAMES')
sns.barplot(x='Rating',y='name',data=low_per_rat)
```

```python
#top_director=movies_df.groupby('director')['name'].count().sort_values(ascending = False).head(10)
top_director=movies_df.director.value_counts().head(10)
top_director
```

```python
max_movies=movies_df.year.value_counts().head(10)
```

```python
max_movies
```

```python
plt.title("MAXIMUM GAMES in YEAR")
sns.barplot(x=max_movies,y=max_movies.index,palette = 'Spectral');
plt.ylabel("NO OF GAMES")
```

```python
most_by_genre = movies_df.groupby('genre')['name'].count().sort_values(ascending = False).head(10)
most_by_genre
```

```python
plt.title('TOP GENRES')
sns.barplot(x=most_by_genre,y=most_by_genre.index)
```
